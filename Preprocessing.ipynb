```python
### MENGAMBIL 30% dari DATASET
# Import Library

from pathlib import Path
from os import listdir
from os.path import isfile, join
from random import sample
import shutil
import os
```

```python
# menampung directory kedalam var 

dataset_dir_default = Path("Rice_Default_DS/")  #awal 
mini_dataset_dir = Path("Rice") #tujuan
```

```python
# melakukan list buat cari tau ada bbrp folder di dalam folder Rocks

onlyfolder = [f for f in listdir(dataset_dir_default) if not isfile(join(dataset_dir_default,f))]
onlyfolder[:10]
```

```python
Path(dataset_dir_default,onlyfolder[0])
```

```python
#buat melakukan cek terhadap file

onlyfiles = [f for f in listdir(Path(dataset_dir_default,onlyfolder[0])) if isfile(join(Path(dataset_dir_default,onlyfolder[0]),f))]
onlyfiles[:10]
```

```python
#buat melakukan random sample 

list_mini_ds = sample(onlyfiles,round(.2*len(onlyfiles)))
list_mini_ds
```

```python
# Buat directory untuk menampung data test

os.mkdir("Rice")
```

```python
# start Processing

for folder in onlyfolder :
    # buat folder di folder Rocks-test
    os.mkdir(Path(mini_dataset_dir,folder))
    
    # ambil 30% dari data
    # 1. buat list semua file di folder tersebut
    onlyfiles = [f for f in listdir(Path(dataset_dir_default,folder)) if isfile(join(Path(dataset_dir_default,folder),f))]
    
    #2. ambil 30% dari data
    list_mini_ds = sample(onlyfiles,round(.3*len(onlyfiles)))
    
    #3. Pindahkan semua file di list_moved
    for filename in list_mini_ds :
        shutil.move(Path(dataset_dir_default,folder,filename), Path(mini_dataset_dir,folder,filename))
```
